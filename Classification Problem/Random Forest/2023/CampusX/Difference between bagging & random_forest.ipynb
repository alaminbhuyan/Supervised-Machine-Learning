There are some difference between Bagging and Random forest. First of all in bagging we can use different base algorithm. That measn we can use all base algorithm such as KNN or SVC or DT. If we use KNN as a base algorithm then all base algorithm will be KNN. But in random forest we have to use as a base algorithm only decision tree. Now, if someone say if I use decision tree as a base algorithm in bagging technique then can I say this is same as random forest. The answer is no. Because, still there are some difference. That difference is in bagging technique if we use decision tree as a base algorithm, when we will give sample data every decision tree the decision tree will be split based on these particular column or row. I think this is confusing. Okay, let me give you a very good example, Look above I draw two image diagram. In bagging image, let's consider first decision tree for our example. We we will give 2 and 4 column by column sampling then decsion tree will be split based on the 2 and 4 column. That mean in bagging spliting decide in tree label. On the other hand random forest if you give 2 and 4 column to the first decision tree then first node will be split based on 2 and 4 column. When second node will be split in that time it will do again column sampling, suppose that time second node split base on 1 and 7 column. Here, I just take random column number for my example. In short, in random forest we bring more randomness so random forest sometimes do well than bagging.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_features=5, n_redundant=0, n_informative=5, n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])

df['target'] = y

print(df.shape)

df.head()
```

```python
# max_features = 2 means we just take two column we do random sampling

bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(X=df.iloc[:,:5], y=df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))

plot_tree(bag.estimators_[0])

plt.show()

# as you can every node split based on either 0 or 1 column
```

```python
# max_features = 2 means we just take two column we do random sampling

rf = RandomForestClassifier(max_features=2)
```

```python
rf.fit(X=df.iloc[:,:5], y=df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[4])
plt.show()
```

```python
# Here as you can see above tree every node split based on different column
```
